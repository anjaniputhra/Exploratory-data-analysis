```python
#distribution of accidents along 24 hrs of a day

req_file.Start_Time = pd.to_datetime(req_file.Start_Time)
```

```python
sbn.distplot(req_file.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
#distribution of accidents along 7 days a week

sbn.distplot(req_file.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

```python
#distribution plot for accidents happening on sunday and monday

sundays_start_time = req_file.Start_Time[req_file.Start_Time.dt.dayofweek == 6]
sbn.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
monday_start_time = req_file.Start_Time[req_file.Start_Time.dt.dayofweek == 0]
sbn.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
#distribution plot of accidents happening every month in 2019

req_file_2019 = req_file[req_file.Start_Time.dt.year == 2019]
sbn.distplot(req_file.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

```python
#using folium to plot a heat map

sample_df = req_file.sample(int(0.001 * len(req_file)))
lat_lon_pairs = list(zip(list(sample_df.Start_Lat), list(sample_df.Start_Lng)))

map = folium.Map()
HeatMap(lat_lon_pairs).add_to(map)
map
```
